```python
import pandas as pd
```

```python
df_net = pd.read_csv('netflix_titles.csv')
df_net
```

```python
df_imdb = pd.read_csv('imdb.csv')
df_imdb = df_imdb.drop(['Poster_Link', 'Gross', 'No_of_Votes'], axis=1)
cast = df_imdb['Star1'] +', ' + df_imdb['Star2'] +', ' + df_imdb['Star3'] +', ' + df_imdb['Star4']
pd.DataFrame(cast)
df_imdb['cast'] = cast
df_imdb.drop(['Star1', 'Star2', 'Star3', 'Star4'], axis=1)
```

```python
df_rot = pd.read_csv('rot_tmt.csv')
df_rot = df_rot[['movie_title', 'rotten_tomatoes_link', 'directors', 'authors', 'original_release_date', 'genres']]
df_rot
```

```python
dataset = pd.DataFrame()

cast = df_net['cast'] + df_imdb['cast'] + df_rot['authors']
dataset['Cast'] = cast
dataset.dropna(inplace=True)

all_casts = dataset['Cast'].str.split(',')

flattened_casts = [name.strip() for sublist in all_casts for name in sublist]

name_counts = pd.Series(flattened_casts).value_counts()

repeated_names = name_counts[name_counts > 1]

def remove_duplicates(cast_list):
    if isinstance(cast_list, list):
        return ', '.join(sorted(set(name.strip() for name in cast_list)))
    return None

dataset['Unique_Cast'] = dataset['Cast'].str.split(',').apply(remove_duplicates)
dataset
#ta tudo errado
```
